```python
plt.figure(figsize=(12,5))
sns.boxplot(data=data[numerical_cols])
plt.title("Boxplot of Scores (Check for Outliers)")
plt.show()
```

```python
numerical_cols = ['math_score', 'reading_score', 'writing_score']

for col in numerical_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)][col]
    
    print(f"\nOutliers in {col}:")
    print(outliers.values)
```

```python
from scipy.stats import zscore

numerical_cols = ['math_score', 'reading_score', 'writing_score']

# Applying z-score
data_zscore = data.copy()
data_zscore[numerical_cols] = data_zscore[numerical_cols].apply(zscore)

# Check
print(data_zscore.head())
```
